```python
# import packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException
import time 
import os
import json
import urllib
```

```python
# Some urls are blocked from untrusted sites, the below code make python environment to trust every site
import os, ssl
if (not os.environ.get('PYTHONHTTPSVERIFY', '') and
getattr(ssl, '_create_unverified_context', None)):
    ssl._create_default_https_context = ssl._create_unverified_context
```

```python
def download_images():
    
    f = open('site_urls.txt','r')
    search_url = f.readlines()[11]
    print(search_url)
   
    # path of chorme drive to run selenium
    path = r'chromedriver'
    
    driver = webdriver.Chrome(path)
    
    k = open('wildlife keywords.txt','r')
    kw = k.readlines()
    
    for i in range(0,len(kw)):
        print(i)
        keyW = kw[i]
        print(keyW)
        time.sleep(4)
        driver.get(search_url)
        # X mark clicking in search text box
        driver.find_element_by_xpath('//*[@id="gnav-search"]/div/div[1]/button[1]').click()
        time.sleep(4)
        # search text box 
        box = driver.find_element_by_xpath('//*[@id="global-enhancements-search-query"]')  
        #sending the keyword
        box.send_keys(keyW)
        time.sleep(6)
        
        # try block if there are no ad results for the keyword it will continue with next keyword
        try:
        #links = driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/div[4]/div[5]/div[3]/div[9]')
        #sub_lnks = links.find_elements_by_xpath('a')

            lnks_lst = [] 
            # collecting links of ad results
            for m in range(1,51):
                lnks_lst.append(driver.find_element_by_xpath('//*[@id="content"]/div/div[1]/div/div[4]/div[5]/div[3]/div[9]/div[1]/div/div/ul/li['+str(m)+']/div/div/a').get_attribute('href'))
            print(lnks_lst[0:3])
        except:
            pass
        

        try:
            # making a directory like URL+Keyword_number , example 11_0 , 11_1,...
            if lnks_lst:
                if not os.path.exists(str(11)+'_'+str(i)):
                    os.mkdir(str(11)+'_'+str(i))
            #  iterating through ads results 
            for j in range(1,3):
                print(lnks_lst[j])
                driver.get(lnks_lst[j])
                time.sleep(2)
                # Title capturing 
                sub_txt = driver.find_element_by_xpath('//*[@id="listing-page-cart"]/div[2]/h1').text
                #making directory inside the above created directory like 0 means first ad, 1 for second ad....
                if not os.path.exists(str(11)+'_'+str(i) +'/'+ str(j)):
                    os.mkdir(str(11)+'_'+str(i) +'/'+ str(j))
                # making dictionary
                details = {
                    'name' : sub_txt,
                    'price' : driver.find_element_by_xpath('//*[@id="listing-page-cart"]/div[3]/div[1]/div[1]/div[1]/div/div/div[1]/p').text,
                    'location': driver.find_element_by_xpath('//*[@id="shipping-variant-div"]/div/div[2]/div[9]').text,
                    'details': driver.find_element_by_xpath('//*[@id="product-details-content-toggle"]/div/ul').get_attribute("innerText"),
                    'description' : driver.find_element_by_xpath('//*[@id="wt-content-toggle-product-details-read-more"]/p').get_attribute("innerText"),
                    'url' : lnks_lst[j]
                }
                # writing the dictionary to text file like 0.txt for first ad ....
                with open(str(11)+'_'+str(i) +'/'+ str(j) + '/' + str(j) + '.txt', 'w') as tfile:
                     tfile.write(json.dumps(details))
                # collecting image urls
                image_lst = []
                elem1 = driver.find_element_by_xpath('//*[@id="listing-right-column"]/div/div[1]/div[1]/div/div/div[2]/div/div[2]/div[3]/ul')
                sub = elem1.find_elements_by_tag_name('img')
                for k,l in enumerate(sub):
                    src = l.get_attribute('src')
                    print(src)
                    image_lst.append(src)
                    # collecting images and saving it inside the above created directory like 0_0.jpg for first ad first image
                    # 0_1.jpg for first ad second image ...
                    if src != None:
                        urllib.request.urlretrieve(str(src), os.path.join(str(11)+'_'+str(i), str(j), str(j)+'_'+str(k)+'.jpg'))
                # writing image urls to text file like 0_imgs.txt for first ad ....
                with open(str(11)+'_'+str(i) +'/'+ str(j) + '/' + str(j) + '_imgs.txt', 'w') as tfile:
                         tfile.write(json.dumps(image_lst)) 
        
        except:
            continue
    driver.close()
    
```

```python
def main():
    download_images()
    
if __name__ == "__main__":
    main()
    
```
