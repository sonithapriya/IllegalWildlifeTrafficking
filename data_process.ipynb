```python
import os
from PIL import Image 
import PIL
import json

path = "data_filter"

lines = ['image,caption']
count = 0
for direc in os.listdir(path):
    if not direc.startswith('.'):
        #print(direc)
        count= count+1
        print(count)
        #try:
        for files in os.listdir(os.path.join(path,direc)):
            if not files.startswith('.'):
                print(path+"/"+direc+"/"+files+"/"+files+".txt")
                f = open(path+"/"+direc+"/"+files+"/"+files+".txt",'r')
                cap = json.load(f)["name"]
                for file in os.listdir(os.path.join(path,direc,files)):
                    if not file.startswith('.'):
                        #print(file)
                        if (file.endswith(".jpg")):
                            #print(file)
                            img = Image.open(os.path.join(path,direc,files,file)) 
                            img.save("data_process/Images/"+direc+"_"+file)
                            lines.append(direc+"_"+file+","+cap)
        #except:
            #pass
    
with open('data_fil_Pro/captions.txt', 'w') as f:
    for line in lines:
        f.write(line)
        f.write('\n')
```
